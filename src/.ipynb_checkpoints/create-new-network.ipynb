```python
import pandas as pd
import numpy as np
import networkx as nx
```

```python
df = pd.read_csv("/Users/jennie/Desktop/INFO6850/networks-paper-1/scratch/opp_migration_with_features.csv")
od_pooled = pd.read_csv("/Users/jennie/Desktop/INFO6850/networks-paper-1/scratch/od_pooled.csv")
```

```python
df['o_cz'] = df['o_cz'].astype(str)
df['d_cz'] = df['d_cz'].astype(str)

df['o_cz_name'] = od_pooled['o_cz_name']
df['d_cz_name'] = od_pooled['d_cz_name']
df['o_state_name'] = od_pooled['o_state_name']
df['d_state_name'] = od_pooled['d_state_name']
```

```python
p = []
p_ = []
for i in df.columns:
    if "o_" in i:
        p.append(i)
    elif "d_" in i:
        p_.append(i)
```

```python
o = df[p].drop_duplicates()
o.columns = o.columns.str.replace("o_",'')
d = df[p_].drop_duplicates()
d.columns = d.columns.str.replace("d_","")
o = o.drop(['d_distance'], axis=1)

o = o.drop_duplicates()
d = d.rename(columns={'frac_between_p25_anp75':'frac_between_p25_and_p75', 'householincome_per_capita':'household_income_per_capita'})
```

```python
node_list = pd.concat([o,d],axis=0).drop_duplicates().set_index('cz')
node_list
```

```python
node_list = node_list.to_dict('index')
```

```python
print("Importing migration data ...")
migration_net = nx.from_pandas_edgelist(
    df=df,
    source="o_cz",
    target="d_cz",
    edge_attr=True,
    create_using=nx.DiGraph
)

# print("Generating quick viz ...")
# pos = nx.spring_layout(G=migration_net, weight='n')
# nx.draw_networkx_nodes(migration_net, pos=pos)

print("Outputting network data ...")
nx.write_graphml(migration_net, "/Users/jennie/Desktop/INFO6850/networks-paper-1/scratch/migration_network.graphml")
```
